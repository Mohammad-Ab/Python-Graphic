```python
#First Ex
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('trex.png')

FlipVertical = cv2.flip(img,0)
FlipHorizontal = cv2.flip(img,1)
FlipBoth = cv2.flip(img,-1)

#making same size
height = 300
width1 = int(img.shape[1] * height / img.shape[0])
width2 = int(FlipHorizontal.shape[1] * height / FlipHorizontal.shape[0])
width3 = int(FlipVertical.shape[1] * height / FlipVertical.shape[0])
width4 = int(FlipBoth.shape[1] * height / FlipBoth.shape[0])
img1 = cv2.resize(img, (width1, height))
img2 = cv2.resize(FlipHorizontal, (width2, height))
img3 = cv2.resize(FlipVertical, (width3, height))
img4 = cv2.resize(FlipBoth, (width4, height))

# Horizontally concatenate the four images
combined_img_1_2 = cv2.hconcat([img1, img2])
combined_img_3_4 = cv2.hconcat([img3, img4])
# Vertically concatenate the two pairs of images
combined_img_all = cv2.vconcat([combined_img_1_2, combined_img_3_4])


cv2_imshow(combined_img_all)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Second Ex
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the three input images
img1 = cv2.imread('beach.png')
img2 = cv2.imread('trex.png')
img3 = cv2.imread('coins.png')


#resizing image
width=250
height=250
dim=(width, height)
re_img1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
re_img2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
re_img3 = cv2.resize(img3, dim, interpolation = cv2.INTER_AREA)

# Crop 1/3 of each image from the right side
height = re_img1.shape[0]
width = int(re_img1.shape[1] / 3)
img1_croped = re_img1[:, :width]
img2_croped = re_img2[:, :width]
img3_croped = re_img3[:, :width]

# Horizontally concatenate the three images
combined_img = cv2.hconcat([img1_croped,img2_croped,img3_croped])

# Display the combined image
cv2_imshow(combined_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Third Ex
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#load the input image
img = cv2.imread('trex.png')

#make even rows red
img[::2, :, 0] = 0
img[::2, :, 1] = 0
img[::2, :, 2] = 255

#make multiple of five rows green
img[::5, :, 0] = 0
img[::5, :, 1] = 255
img[::5, :, 2] = 0

#make multiple of seven rows white
img[::7, :, 0] = 255
img[::7, :, 1] = 255
img[::7, :, 2] = 255

#display the output image
cv2_imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Fourth Ex
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Load the input image
org_img = cv2.imread('beach.png')
img = cv2.imread('beach.png')

# Add 100 to the red channel
img[:, :, 2] = cv2.add(img[:, :, 2], 100)

# Subtract 50 from the green channel
img[:, :, 1] = cv2.subtract(img[:, :, 1], 50)

# Double the blue channel
img[:, :, 0] = cv2.multiply(img[:, :, 0], 2)

# Create subplots for original and modified images
fig, axs = plt.subplots(1, 2)
axs[0].imshow(cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Modified Image')

plt.show()
```

```python
#Fifth Ex
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the input image
img = cv2.imread('beach.png', cv2.IMREAD_GRAYSCALE)

# Divide the image into 4 blocks
height, width = img.shape
block_height = height // 2
block_width = width // 2
block_A = img[:block_height, :block_width]
block_B = img[:block_height, block_width:]
block_C = img[block_height:, block_width:]
block_D = img[block_height:, :block_width]

# Change the order of the blocks to B, C, D and A
img_out = cv2.hconcat([cv2.vconcat([block_B, block_C]), cv2.vconcat([block_D, block_A])])

# Display the output image
cv2_imshow(img_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Sixth Ex
import cv2
from google.colab.patches import cv2_imshow

# Load the input image
img = cv2.imread('trex.png', cv2.IMREAD_GRAYSCALE)

# Apply a color map based on the pixel intensity
img_out = cv2.applyColorMap(img, cv2.COLORMAP_JET)
img_out[img <= 100] = [0, 0, 255]    # Red
img_out[(img > 100) & (img <= 150)] = [0, 128, 128]    # Brown
img_out[(img > 150) & (img <= 200)] = [0, 255, 0]    # Green
img_out[img > 200] = [0, 255, 255]    # Yellow

# Display the output image
cv2_imshow(img_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Seventh Ex
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Load the input images
img1 = cv2.imread('trex.png')
img2 = cv2.imread('beach.png')
#resizing image
width=233
height=350
dim=(width, height)
re_img1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
re_img2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
# Create subplots for original images
fig, axs = plt.subplots(3, 2)
axs[0, 0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Image 1')
axs[0, 1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('Image 2')

# Set channel of second image instead of first image
img_out_set = re_img1.copy()
for i in range(3):
  img_out_set[:, :, i] = re_img2[:, :, i]
  if i == 0:
    axs[1, 0].imshow(cv2.cvtColor(img_out_set, cv2.COLOR_BGR2RGB))
    axs[1, 0].set_title('Step 1')
  elif i == 1:
    axs[1, 1].imshow(cv2.cvtColor(img_out_set, cv2.COLOR_BGR2RGB))
    axs[1, 1].set_title('Step 2')
  else:
    axs[2, 0].imshow(cv2.cvtColor(img_out_set, cv2.COLOR_BGR2RGB))
    axs[2,0].set_title('Step 3')

axs[2,1].imshow(cv2.cvtColor(img_out_set, cv2.COLOR_BGR2RGB))
axs[2,1].set_title('final image')


fig.suptitle('Final Result')
axs[0, 0].axis('off')
axs[0, 1].axis('off')
axs[1, 0].axis('off')
axs[1, 1].axis('off')
axs[0, 0].set_aspect('equal')
axs[0, 1].set_aspect('equal')
axs[1, 0].set_aspect('equal')
axs[1, 1].set_aspect('equal')

plt.show()
```

```python
#Eighth Ex
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the input image
img = cv2.imread('trex.png')

# Take two numbers from the user for rows
row1 = int(input('Enter first row number: '))
row2 = int(input('Enter second row number: '))

# Get the pixel values for the specified rows
row1_pixels = img[row1, :, :]
row2_pixels = img[row2, :, :]

# Draw histograms of the three channel numbers that the user enters
for i in range(3):
    plt.hist(row1_pixels[:, i], bins=256, color='red', alpha=0.5)
    plt.hist(row2_pixels[:, i], bins=256, color='blue', alpha=0.5)
    plt.xlim([0, 256])
    plt.title('Histogram of Row {} and Row {}'.format(row1, row2))
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Ninth Ex
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load the input image
org_img = cv2.imread('trex.png')
img = cv2.imread('trex.png')

# Set salt and pepper noise
noise = np.zeros(img.shape, np.uint8)
cv2.randu(noise, 0, 255)
black = noise < 30
white = noise > 225
img[black] = 0
img[white] = 255

# Define the kernel for the filter2D function
kernel = np.ones((5,5),np.float32)/25

# Apply the filter2D function to blur the image
img_blur = cv2.filter2D(img,-1,kernel)

# Display the original and blurred images side by side
cv2_imshow(org_img)
cv2_imshow(img_blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
