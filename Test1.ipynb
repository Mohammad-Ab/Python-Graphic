```python
from matplotlib import pyplot as plt
import numpy as np

array = np.random.randint(20, size=(8, 8))
print(array)
k=0
list1=[]
x=[]
#make a list of double blocks
for i in range (8):
  for j in range(8):
    x.append(array[i][j])
    k+=1
    if k == 2:
      list1.append(x)
      x=[]
      k=0
print("------------------------------------------")
#convert our double block list to an array
newarray = np.array(list1)
print("NEW ARRAY OF DOBLE LIST")
print(newarray)
#making a min and max list
minlist = newarray.min(axis=1)
maxlist = newarray.max(axis=1)

print("THE MIN LIST IS:")
print(minlist)
print("THE MAX LIST IS:")
print(maxlist)
print("-----------------------------------------------------")
#now draw min and max list by matplotlib
plt.plot(minlist, 'o-r')
plt.title("min of the blocks")
plt.show()
print("------------------------------------------------------")
plt.plot(maxlist, 'o-b')
plt.title("max of the blocks")
plt.show()
print("------------------------------------------------------")
#use a dictionary for count the numbers
checkdict={'bad':0, 'medium':0, 'good':0, 'best':0 }

for i in range(32):
  if minlist[i] and maxlist[i] in range(0,10):
    checkdict['bad']+=1
  elif minlist[i] and maxlist[i] in range(10,15):
    checkdict['medium']+=1
  elif minlist[i] and maxlist[i] in range(15,18):
    checkdict['good']+=1
  else:
    checkdict['best']+=1

print(checkdict)
```
